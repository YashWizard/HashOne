```python
import numpy as np
import pandas as pd

# Load the train data (as in your previous setup)
npz_file = np.load('/kaggle/input/chbmitseizuredataset/eeg-seizure_train.npz', allow_pickle=True)
signals_train = npz_file['train_signals']
labels_train = npz_file['train_labels']

# Reshape the signals array into 2D: (num_samples, 23*256)
num_samples = signals_train.shape[0]
reshaped_signals = signals_train.reshape(num_samples, -1)  # Flatten each sample to 1D

# Create column names for the DataFrame, like 'Channel_0_Point_0', 'Channel_0_Point_1', ..., 'Channel_22_Point_255'
columns = [f'Channel_{ch}_Point_{pt}' for ch in range(23) for pt in range(256)]

# Create DataFrame
df = pd.DataFrame(reshaped_signals, columns=columns)

# Add labels column to the DataFrame (optional, if you want to include labels)
df['label'] = labels_train

# Display the DataFrame
print(df.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft

# Assuming 'signals_train' has already been loaded and has shape (nr_samples, 23, 256)

# Select a single sample and one channel (e.g., first sample, first channel)
sample_idx = 0  # Pick any sample index
channel_idx = 0  # Pick the first channel (or any channel you want)

eeg_signal = signals_train[sample_idx, channel_idx, :]  # Shape is (256,)

# Perform FFT on the EEG signal
fft_signal = fft(eeg_signal)

# Get the amplitude spectrum (magnitude of FFT)
amplitude = np.abs(fft_signal)

# Generate the frequency axis
sampling_rate = 256  # Hz
freq = np.fft.fftfreq(len(eeg_signal), d=1/sampling_rate)

# We only care about the positive frequencies (first half of the spectrum)
half_n = len(freq) // 2
freq = freq[:half_n]
amplitude = amplitude[:half_n]

# Plot frequency vs amplitude
plt.figure(figsize=(10, 6))
plt.plot(freq, amplitude, color='b')
plt.title(f'Frequency vs Amplitude for Sample {sample_idx}, Channel {channel_idx}')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'signals_train' is already loaded with shape (nr_samples, 23, 256)

# Set the number of samples to plot (first 5 samples)
num_samples_to_plot = 5
channel_idx = 0  # Plot for Channel 0 (can change to any other channel)

# Generate time array: from 0 to 1 second with 256 points (sampling rate is 256 Hz)
sampling_rate = 256  # Hz
time = np.linspace(0, 1, sampling_rate)

# Plot each sample as a line
plt.figure(figsize=(10, 6))
for i in range(num_samples_to_plot):
    eeg_signal = signals_train[i, channel_idx, :]  # Select the i-th sample, Channel 0
    plt.plot(time, eeg_signal, label=f'Sample {i}')

# Set plot labels and title
plt.title(f'EEG Signals for First {num_samples_to_plot} Samples (Channel {channel_idx})')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude (µV)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
