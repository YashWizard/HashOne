```python
import cv2
import mediapipe as mp
import pyautogui
import time
```

```python
# Initialize MediaPipe Hands
capture_hands = mp.solutions.hands.Hands(static_image_mode=False, max_num_hands=1, min_detection_confidence=0.7, min_tracking_confidence=0.7)
drawing_option = mp.solutions.drawing_utils
screen_width, screen_height = pyautogui.size()

# Set up video capture
camera = cv2.VideoCapture(0)
camera.set(cv2.CAP_PROP_FPS, 60)  # Set camera FPS
camera.set(cv2.CAP_PROP_FRAME_WIDTH, 520)  # Reduce frame size for faster processing
camera.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

x1 = y1 = x2 = y2 = 0
frame_counter = 0

while True:
    start_time = time.time()
    ret, image = camera.read()
    if not ret:
        break

    image = cv2.flip(image, 1)
    image_height, image_width, _ = image.shape
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Limit processing to every 3rd frame
    frame_counter += 1
    if frame_counter % 3 == 0:
        output_hands = capture_hands.process(rgb_image)
        all_hands = output_hands.multi_hand_landmarks
        if all_hands:
            for hand in all_hands:
                drawing_option.draw_landmarks(image, hand)
                one_hand_landmarks = hand.landmark
                for id, lm in enumerate(one_hand_landmarks):
                    x = int(lm.x * image_width)
                    y = int(lm.y * image_height)
                    if id == 8:
                        mouse_x = int(screen_width / image_width * x)
                        mouse_y = int(screen_height / image_height * y)
                        cv2.circle(image, (x, y), 10, (0, 255, 255), -1)
                        pyautogui.moveTo(mouse_x, mouse_y)
                        x1 = x
                        y1 = y
                    if id == 4:
                        x2 = x
                        y2 = y
                        cv2.circle(image, (x, y), 10, (0, 255, 255), -1)
            dist = y2 - y1
            if dist < 20:
                pyautogui.click()

    # Display the frame
    cv2.imshow("Hand movement video capture", image)
    end_time = time.time()
    fps = 1 / (end_time - start_time)
    print(f"FPS: {fps:.2f}")
    
    key = cv2.waitKey(1)
    if key == 27:  # Press 'ESC' to quit
        break

camera.release()
cv2.destroyAllWindows()
```
