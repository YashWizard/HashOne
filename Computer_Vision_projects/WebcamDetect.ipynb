## Import dependencies

```python
## Import opencv for computer vision stuff.....
import cv2
## Import matplotlib so we can visualize an image....
from matplotlib import pyplot as plt
```

## Connecting to our webcam

```python
cap = cv2.VideoCapture(0)
```

```python
ret, frame = cap.read()
```

```python
plt.imshow(frame)
```

```python
## release capture into the wild
cap.release()
```

```python
def take_photo():
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    cv2.imwrite('webcamphoto.jpg', frame)
    cap.release()
```

```python
take_photo()
```

## Rendering in Real time

```python
## Connect to our webcam first !!!
cap = cv2.VideoCapture(0)

## Establishes a loop through every frame until we close the video capture
while cap.isOpened():
    ret, frame = cap.read()

    ## Show image
    cv2.imshow('Webcam', frame)

    ##Check whether q has been hit and stops the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

#Release the webcam
cap.release()
#Close the frame
cv2.destroyAllWindows()
```
